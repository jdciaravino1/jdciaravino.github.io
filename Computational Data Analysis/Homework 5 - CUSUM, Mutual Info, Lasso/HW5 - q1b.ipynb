```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
random.seed(10)
mean0 = 0
sigma0 = 1
out0 = np.random.normal(mean0, sigma0, 100)
```

```python
x = np.linspace(0,5,10,endpoint=False)
x
```

```python
out0
```

```python
mean1 = 0.5
sigma1 = 1
out1 = np.random.normal(mean1, sigma1, 100)
```

```python
out1
```

```python
full_set = np.append(out0,out1)
```

```python
len(full_set)
```

```python
w = 0
w_list = []
for i in full_set:
    f0 = multivariate_normal.pdf(i,mean=0,cov=1)
    f1 = multivariate_normal.pdf(i,mean=0.5,cov=1)
    log_ratio = math.log2(f1/f0)
    w = max(w+log_ratio,0)
    w_list.append(w)
    #w = max(w+math.log2())
```

```python
len(w_list)
```

```python
test = list(range(1,201))
test
len(test)
```

```python
# plot
fig, ax = plt.subplots(dpi = 500)

ax.plot(list(range(1,201)), w_list, linewidth=1.0)

ax.set(xlim=(0, 201), xticks=np.arange(0, 201),
       ylim=(0, 20), yticks=np.arange(0, 20))

# plotting a line plot after changing it's width and height
plt.figure(figsize=(200,200)) 

plt.show()
```
